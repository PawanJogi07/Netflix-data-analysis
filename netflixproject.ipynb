```python
import pandas as pd
import matplotlib.pyplot as  plt
df= pd.read_csv('netflix_titles.csv')
print (df.head())
df = df.dropna(subset = ['type','release_year','rating','country','duration'])
type_counts = df['type'].value_counts()
plt.figure(figsize= (6,4))
plt.bar(type_counts.index,type_counts.values,color = ('skyblue','orange'))
plt.title('movies vs tv shows')
plt.xlabel ('type')
plt.ylabel('count')
plt.tight_layout()
plt.savefig('movies or tv shows.png',dpi=300,bbox_inches='tight')
plt.show()
  
```

```python

rating_counts = df['rating'].value_counts()
plt.figure(figsize = (8,6))
plt.pie(rating_counts, labels=rating_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Percentage of Ratings')
plt.tight_layout()
plt.savefig('movies_rating.png', dpi=300, bbox_inches='tight') 
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as  plt
df= pd.read_csv('netflix_titles.csv')
print (df.head())
df = df.dropna(subset = ['type','release_year','rating','country','duration'])
movies_df = df[df['type'] == 'movie'].copy()
movies_df['duration_int'] = movies_df['duration'].str.replace('min', '').astype(int)
plt.figure(figsize=(8,6))
plt.hist(movies_df['duration_int'],bins = 30,coloredge= 'green')
plt.xlable('movieduration')
plt.ylabel('movierating')
plt.savefig('movieis.png')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("netflix_titles.csv")
df = df.dropna(subset = ['type','release_year','country','rating','duration'])
content_by_year = df.groupby(['release_year','type']).size().unstack().fillna(0)
fig,ax=plt.subplots(1,2,figsize= (12,10))
ax[0].plot(content_by_year.index, content_by_year['Movie'], color='blue')
ax[0].set_title('movies')
ax[0].set_xlabel('mvies_year')
ax[0].set_ylabel('content')

ax[0].plot(content_by_year.index, content_by_year['TV Show'], color='orange')
ax[0].set_title('tvshow')
ax[0].set_xlabel('tv show')
ax[0].set_ylabel('content')
fig.suptitle('comparision')
plt.tight_layout()
plt.savefig('comparision.png',dpi=300,bbox_inches='tight')
plt.show()
```
